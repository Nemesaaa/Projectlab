```python
def mean_list(list_of_lists):
    mean_list = []
    sd_list = []
    values_list = []

    for j in range(len(list_of_lists[0])):
        current_values =[]
        for i in list_of_lists:
            current_values.append(i[j])
        values_list.append(current_values)
    print(values_list)
    for i in values_list:
        mean_list.append(sum(i) / len(i))
        sd_list.append(statistics.pstdev(i))

    mean_plus_sd_list = np.add(mean_list,sd_list)
    mean_minus_sd_list = np.subtract(mean_list,sd_list)

    return mean_list, mean_plus_sd_list, mean_minus_sd_list
```

```python
def double_sd_list(list_of_lists):
    mean_list = []
    sd_list = []
    values_list = []

    for j in range(len(list_of_lists[0])):
        current_values =[]
        for i in list_of_lists:
            current_values.append(i[j])
        values_list.append(current_values)
    print(values_list)
    for i in values_list:
        mean_list.append(sum(i) / len(i))
        sd_list.append(statistics.pstdev(i))

    mean_plus_sd_list = np.add(mean_list,sd_list)
    mean_plus_sd_list = np.add(mean_plus_sd_list,sd_list)
    mean_minus_sd_list = np.subtract(mean_list,sd_list)
    mean_minus_sd_list = np.subtract(mean_minus_sd_list,sd_list)

    return mean_plus_sd_list, mean_minus_sd_list
```
